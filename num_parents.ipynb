## __Number of Parents__

Let's say we have some population of $N$ individuals.

Each generation, we will select $p$ parents from the population.

If each selected parent mates with every other selected parent once,  

there will be $\displaystyle \binom{p}{2}$ unique pairings.

Let's say we want a total of $n$ new children to be born each generation,  
and each mating pair will produce exactly $b$ babies,  
then we can calculate the number of parents necessary, $p$:

$$
\begin{align*}
n &= \binom{p}{2} \cdot b \\ \\
\frac{n}{b} &= \frac{p!}{((p-2)! \cdot 2!)} \\ \\
\frac{2n}{b} &= \frac{p!}{(p-2)!} \\ \\
\frac{2n}{b} &= \frac{p(p-1)(p-2)!}{(p-2)!} \\ \\
\frac{2n}{b} &= p(p-1) \\ \\
\Rightarrow p^2 - p - \frac{2n}{b} &= 0, \quad \text{where} \ p \ge 0 \\ \\

\end{align*}
$$

We can use the quadratic formula to solve for the roots:

$$
\begin{align*}
p &= \frac{ -(-1) \pm \sqrt{(-1)^2 - 4(1)\left(-\displaystyle \frac{2n}{b}\right)}} {2(1)} \\ \\
p &= \frac{1 \pm \sqrt{1 + \displaystyle \frac{8n}{b}}}{2}
\end{align*}
$$

We will ignore the negative root since $p \ge 0$.  
Since, $p$ must be a whole number, we will also round up $p$ to the nearest integer.

$$
\therefore \quad \left\lceil p = \frac{1 + \sqrt{1 + \displaystyle \frac{8n}{b}}}{2} \right\rceil
$$

## __Let's graph the relationship!__

```python
import math
import numpy as np
import matplotlib.pyplot as plt

def f(n, b):
	''' Returns number of parents required given:
		- each parent mates with another once
		- n children must be born total
		- each pair produces b offspring
	'''
	return math.ceil((1 + math.sqrt(1 + (8*n/b))) / 2)

# Define the domain for n (birth_rate)
birth_rate = np.linspace(0, 100, 500)  # birth_rate from 0 to 100

# Set up the plot - 1 row, 5 columns
fig, axs = plt.subplots(1, 5, figsize=(20, 5), sharey=True)

# Plots for offspring, b = 1 to b = 5
for i, b in enumerate(range(1, 6)):
	num_parents = [f(n, b) for n in birth_rate]

	axs[i].plot(birth_rate, num_parents, color='darkgreen')
	axs[i].set_title(f'Offspring = {b}')
	axs[i].set_xlabel('Birth Rate (n)')
	axs[i].set_ylabel('Number of Parents (p)')
	axs[i].grid(True, alpha=0.7)

plt.tight_layout()
plt.show()
```
